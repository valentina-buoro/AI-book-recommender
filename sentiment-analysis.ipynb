```python
import numpy as np

emotions_label = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
isbn=[]
emotion_scores = {label: [] for label in emotions_label}

def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotions_label}
    for prediction in predictions:
        sorted_prediction = sorted(prediction, key=lambda x: x["label"])
        for index, label in enumerate(emotions_label):
            per_emotion_scores[label].append(sorted_prediction[index]["score"])
    return {label: np.max(scores) for label, scores in per_emotion_scores.items()}
```

```python
for i in range(10):
    isbn.append(books['isbn13'][i])
    sentences = books['description'][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotions_label:
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm

emotions_label = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
isbn=[]
emotion_scores = {label: [] for label in emotions_label}

for i in tqdm(range(len(books))):
    isbn.append(books['isbn13'][i])
    sentences = books['description'][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotions_label:
        emotion_scores[label].append(max_scores[label])
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
```

```python
emotions_df.head()
```

```python
books = pd.merge(books, emotions_df, on="isbn13")
```
